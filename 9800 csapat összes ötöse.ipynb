```python
import pandas
pbp=pandas.read_csv('./film/hun1920.csv')
```

```python
boxscore=pandas.read_csv('./boxscore/gamehunx1920.csv',header=None)
boxscore.columns=["CODE","team","GM","Min","PF","DF","2Pm","2Pa","2PP","3Pm","3Pa","3PP","FTm","Fta","FTP","ORB","DRB","TRB","BS","SR","TOV","ST","ASS","PTS", "VALU","OP","OER","ASTO","VIR","extra1","date","extra3","nat"]
```

```python
boxscore=boxscore[boxscore['team']==9800]
boxscore=boxscore[boxscore['GM']==1]
playedgames=boxscore['extra1'].unique()
```

```python
starters=dict()
cnt=0
for game in playedgames:
    starter=[]
    for i,row in boxscore.iterrows():
        if((row['team']==9800)&(row['extra1']==game)):
            starter.append(row['CODE'])
    tmp=frozenset(starter)
    if(tmp not in starters):
        starters.update({tmp:cnt})
        cnt+=1
starters
```

```python
pbp['gamecode']=pbp['gamecode'].str.replace('hun_','').astype('int64')
pbp
```

```python
hashtable=dict()
cnt=0
for game in playedgames:
    starter=set()
    for i,row in boxscore.iterrows():
        if((row['team']==9800)&(row['extra1']==game)):
            starter.add(row['CODE'])
    tmp=frozenset(starter)
    if(tmp not in hashtable):
        hashtable.update({tmp:cnt})
        cnt+=1
    pbp_game=pbp[pbp['gamecode']==game].sort_values(by=['sorrend'])
    for i,row in pbp_game.iterrows():
         if((row['teamid']==9800)&(row['fccode']==1011)):
            starter.add(row['playercode'])
            starter.discard(row['playercode2'])
            tmp=frozenset(starter)
            if(tmp not in hashtable):
                hashtable.update({tmp:cnt})
                cnt+=1      
```

```python
hashtable
```

```python
#Összesen 15 játékost szerepeltettek a bajnokságban, ebből 3003 féle ötös variáció lehetséges, de ezek közül csak 129 félét próbáltak ki
```

```python
inorder_hash=[]
played_mins=[]
for game in playedgames:
    lineup=set()
    for i,row in boxscore.iterrows():
        if((row['team']==9800)&(row['extra1']==game)):
            lineup.add(row['CODE'])
    tmp=frozenset(lineup)
    inorder_hash.append(hashtable.get(tmp))
    fr=0
    pbp_game=pbp[pbp['gamecode']==game].sort_values(by=['sorrend'])
    for i,row in pbp_game.iterrows():
         if((row['teamid']==9800)&(row['fccode']==1011)):
            to=int((row['period']-1)*10+row['minute'])
            played_mins.append(to-fr)
            lineup.add(row['playercode'])
            lineup.discard(row['playercode2'])
            tmp=frozenset(lineup)
            inorder_hash.append(hashtable.get(tmp))
            fr=to
    played_mins.append(40-fr)
```

```python
def get_points(row):
    if((row['fccode']==1000)&(row['teamid']==9800)):
            if(row['subcode']==3):
                return 3
            elif(row['subcode']==4):
                return 1
            else:
                return 2
    return 0
```

```python
def get_otherMetrics(row,param,param2=-1):
    if((row['fccode']==param)|(row['fccode']==param2)):
        return 1
    return 0
```

```python
metrics=[]
for game in playedgames:
    points=0
    shoots=0
    missed=0
    defrb=0
    offrb=0
    steal=0
    turnover=0
    assist=0
    block=0
    fd=0
    pbp_game=pbp[pbp['gamecode']==game].sort_values(by=['sorrend'])
    for i,row in pbp_game.iterrows():
        if(row['teamid']==9800):
            points+=get_points(row)
            shoots+=get_otherMetrics(row,1000,1001)
            missed+=get_otherMetrics(row,1001)
            defrb+=get_otherMetrics(row,1002,2002)
            offrb+=get_otherMetrics(row,1003,2003)
            steal+=get_otherMetrics(row,1004,2004)
            turnover+=get_otherMetrics(row,1005,2005)
            assist+=get_otherMetrics(row,1008)
            block+=get_otherMetrics(row,1009)
            fd+=get_otherMetrics(row,1007,2007)
            if(row['fccode']==1011):
                metrics.append([points,shoots,missed,defrb,offrb,steal,turnover,assist,block,fd])
                points=0
                shoots=0
                missed=0
                defrb=0
                offrb=0
                steal=0
                turnover=0
                assist=0
                block=0
                fd=0
    metrics.append([points,shoots,missed,defrb,offrb,steal,turnover,assist,block,fd])  
```

```python
len(metrics)
```

```python
df=pandas.DataFrame(metrics)
df.columns=['points','shots','missed','defrb','offrb','steal','turnover','assist','block','fouldrawn']
df['min']=played_mins
df['hash']=inorder_hash
df.groupby('hash').sum().sort_values(by=['min','points'],ascending=False).head(10)
```

```python
df["VAL"]=df["points"]+df["assist"]+df["defrb"]+df["offrb"]+df["fouldrawn"]+df["steal"]-df["missed"]-df['turnover']+df['block']
df.groupby('hash').sum().sort_values(by=['min','points'],ascending=False).head(20)
```
