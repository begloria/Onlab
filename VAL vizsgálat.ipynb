```python
import pandas
```

```python
li=[]
for i in range(10,20):
    print('{}{}'.format(i,i+1))
    boxscore=pandas.read_csv('./boxscore/gamehunx{}{}.csv'.format(i,i+1),header=None)
    li.append(boxscore)
data = pandas.concat(li, axis=0, ignore_index=True)
```

```python
data.columns=["CODE","team","GM","Min","PF","DF","2Pm","2Pa","2PP","3Pm","3Pa","3PP","FTm","Fta","FTP","ORB","DRB","TRB","BS","SR","TOV","ST","ASS","PTS", "VALU","OP","OER","ASTO","VIR","extra1","date","extra3","nat"]
```

```python
data
```

```python
data["VALU"]=data["PTS"]+data["TRB"]+data["DF"]+data["ST"]+data["ASS"]+data["BS"]-(data["2Pa"]-data["2Pm"])-(data["3Pa"]-data["3Pm"])-(data["Fta"]-data["FTm"])-data['TOV']
```

```python
sum=data.groupby(['extra1','team']).sum().drop(columns=['OP','OER','ASTO','VIR','date','extra3'])
sum
```

```python
sum['2PP']=sum['2Pm']/sum['2Pa']*100
sum['3PP']=sum['3Pm']/sum['3Pa']*100
sum['FTP']=sum['FTm']/sum['Fta']*100
sum
```

```python
sum=sum.reset_index()
teams=sum['team']
teams
```

```python
points=sum['PTS']
val=sum['VALU']
```

```python
cnt=0
for i in range(0,254,2):
    if((points[i]-points[i+1])*(val[i]-val[i+1])<0):
        cnt=cnt+1
cnt
```
