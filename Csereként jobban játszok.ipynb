```python
import pandas
```

```python
data=pandas.read_csv('./boxscore/gamehunx1920.csv',header=None)
```

```python
data.columns=["CODE","team","GM","Min","PF","DF","2Pm","2Pa","2PP","3Pm","3Pa","3PP","FTm","Fta","FTP","ORB","DRB","TRB","BS","SR","TOV","ST","ASS","PTS", "VALU","OP","OER","ASTO","VIR","extra1","date","extra3","nat"]
```

```python
data["VALU"]=data["PTS"]+data["TRB"]+data["DF"]+data["ST"]+data["ASS"]+data["BS"]-(data["2Pa"]-data["2Pm"])-(data["3Pa"]-data["3Pm"])-(data["Fta"]-data["FTm"])-data['TOV']
```

```python
result=data.groupby(['CODE','GM']).sum().drop(columns=['2PP','3PP','OP','OER','ASTO','VIR','date','extra3','extra1','FTP']).reset_index()
```

```python
result['VAL40']=result['VALU']/result['Min']*40
```

```python
from itertools import tee
def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = tee(iterable)
    next(b, None)
    return zip(a, b)
```

```python
players=pandas.read_csv('players_hun1819.csv')
```

```python
for (i1, row1), (i2, row2) in pairwise(result.iterrows()):
    if(row1['CODE']==row2['CODE']):
        if(row1['VAL40']>row2['VAL40']):
            for i,row in players.iterrows():
                if(row['playercode']==row1['CODE']):
                    print(row['playername'],"csereként:",row1['Min'],row1['VAL40'],"kezdőként:",row2['Min'],row2['VAL40'],row2['VAL40']-row1['VAL40'])
```
